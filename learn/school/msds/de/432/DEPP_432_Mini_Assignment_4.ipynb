```python
import numpy as np
import pandas as pd
import time
import math
```

```python
states = ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 
           'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 
           'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 
           'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 
           'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY']
rand_26_idx = np.random.randint(26)
letters = list('abcdefghijklmnopqrstuvwxyz')
```

```python
addresses = list(np.random.choice(states, 50, replace=False).copy())
```

```python
fnames = []
lnames = []
for i in range(50):
    fnames.append(''.join(np.random.choice(letters, 10, replace=False)))
    lnames.append(''.join(np.random.choice(letters, 10, replace=False)))
```

```python
persons = pd.DataFrame(data=[fnames, lnames, addresses]).T
persons.columns = ['fname', 'lname', 'address']
persons_q = persons.copy()
persons_b = persons.copy()
persons_m = persons.copy()
persons_ps = persons.copy()
```

```python
def quicksort(array): 
    if len(array) < 2:
        # Base case: arrays with 0 or 1 element are already “sorted.”
        return array 
    else: 
        # Recursive case
        pivot = array[0]
        # Sub-array of all the elements less than the pivot
        less = [i for i in array[1:] if i <= pivot]
        # Sub-array of all the elements greater than the pivot 
        greater = [i for i in array[1:] if i > pivot] 
        return quicksort(less) + [pivot] + quicksort(greater) 
```

```python
def bubbleSort(alist):
    for passnum in range(len(alist)-1,0,-1):
        for i in range(passnum):
            if alist[i]>alist[i+1]:
                temp = alist[i]
                alist[i] = alist[i+1]
                alist[i+1] = temp
    return alist
```

```python
persons_q.fname = persons.apply(lambda x: quicksort(persons.fname))
```

```python
timer = {'quick':0, 'bubble':0, 'pandas':0}

start_q = time.time()
persons_q.fname = persons.apply(lambda x: quicksort(persons.fname))
end_q = time.time()
time_q = end_q-start_q
timer['quick']=time_q

start_b = time.time()
persons_b.fname = persons.apply(lambda x: bubbleSort(persons.fname))
end_b = time.time()
time_b = end_b-start_b
timer['bubble']=time_b

start_ps = time.time()
persons_ps.sort_values(by='fname')
end_ps = time.time()
time_ps = end_ps-start_ps
timer['pandas']=time_ps

timer = pd.Series(timer)
timer.plot.bar()
```
