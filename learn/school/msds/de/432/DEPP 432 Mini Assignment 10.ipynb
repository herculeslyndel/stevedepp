```python
import numpy as np
import pandas as pd
np.random.seed(seed=42)
```

```python
def top_n(d, n):
    def keyfunction(k):
        return d[k]
    key_list = []
    value_list = []
    for key in sorted(d, key=keyfunction, reverse=False)[:n]:
        key_list.append(key)
        value_list.append(d[key])
    return key_list, value_list
```

```python
def knn_50(k=3):
    buy_bin = np.random.choice([0, 1], size=(50,5))
    buy_bin_df = pd.DataFrame(buy_bin, columns =['Eggs','Bread','Milk','Butter','Soda'])
    shopper_count = 50
    shopper = range(1,shopper_count+1)
    buy_bin_df.insert(0, 'Shopper', shopper, True) 
    knn=[]
    dist=[]
    for neighbor_1 in range(shopper_count):
        dude = buy_bin_df.iloc[neighbor_1, 1:].values
        neighbors = {}
        for neighbor_2 in range(shopper_count):
            dudes_neighbor = buy_bin_df.iloc[neighbor_2, 1:].values
            L2 = round(np.linalg.norm(dude-dudes_neighbor),5)
            neighbors[neighbor_2+1]=L2
            knn_list, knn_dist = top_n(neighbors,k)
        knn.append(knn_list)
        dist.append(knn_dist)
    knn, dist
    buy_bin_df.insert(6, 'knn', knn, True) 
    buy_bin_df.insert(7, 'dist', dist, True)
    return buy_bin_df
```

```python
buy_bin_df = knn_50(4)
buy_bin_df
```

```python
import time
elapsed=[]
for k in range(1,11):
    start = time.time()
    knn_50(k)
    end = time.time()
    print(end-start,k)
    elapsed.append(end-start)
```
