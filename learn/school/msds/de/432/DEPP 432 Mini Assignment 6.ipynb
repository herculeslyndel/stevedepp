```python
import numpy as np
import time
import pandas as pd
import math
import matplotlib.pyplot as plt
from collections import deque, defaultdict
```

```python
fb = []
```

```python
letters = list('abcdefghijklmnopqrstuvwxyz')
```

```python
def get_names(n=5):
    names = []
    for i in range(5):
        names.append(''.join(np.random.choice(letters, 10, replace=False)))
    return names
```

```python
get_names()
```

```python
def bob():
    return 'bob'
graph = defaultdict(bob)
i=0
j=5
k=30
m=155
initial_search_queue=[]
for first_level in get_names():
    i+=1
    fb.append(first_level)
    initial_search_queue.append(first_level)
    graph[first_level] = get_names()
    print(i, 'on 1st=', first_level, j, 'on 2nd =', graph[first_level]) 
    for second_level in graph[first_level]:
        j+=1
        fb.append(second_level)
        graph[second_level] = get_names()
        print(j,'on 2nd=',second_level,j*5+1,'on 3rd:', graph[second_level])    
        for third_level in graph[second_level]:
            k+=1
            fb.append(third_level)
            graph[third_level] = get_names()
            print(k,'on 3rd=',third_level,k*5+1,'on 4th:', graph[third_level])    
            for fourth_level in graph[third_level]:
                m+=1
                fb.append(fourth_level)
                graph[fourth_level] = get_names()
                print(m,'on 4th=',fourth_level,m*5+1,'on 5th:', graph[fourth_level])    
                for fifth_level in graph[fourth_level]:
                    fb.append(fifth_level)
```

```python
initial_search_queue
```

```python
print(len(fb) == len(set(fb)), ', there are no dups in this array of', len(fb), 'names.')
```

```python
def search(dicti, name):
    search_queue = deque()
    search_queue += initial_search_queue
    searched = []
    i=0
    while search_queue:
        i+=1
        person = search_queue.popleft()
        if not person in searched:
            if person==name:
                print (person + 'is found after looking in', i, 'places!')
                return i, True
            else:
                search_queue += graph[person]
                searched.append(person)
    return False
```

```python
level_list = []
looks_list = []
names_list = []
times_list = []
```

```python
first_level = list(graph.keys())[0:5]
search_first_level = np.random.choice(first_level, 1, replace=False)
print("Searching for 1 names from 1st level connections:")
print('Looking for', search_first_level)
for i in search_first_level:
    start = time.time()
    looks = search(graph, i)
    end = time.time()
    elapsed = end-start
    print('Elapsed time is:', elapsed)
    level_list.append(1)
    looks_list.append(looks[0])
    names_list.append(i)
    times_list.append(elapsed)
print(level_list)
```

```python
second_level = list(graph.values())[0]
search_second_level = np.random.choice(second_level, 2, replace=False)
print("Searching for 2 names from same 1st level person's 2nd level connections:")
print('Looking for', search_second_level)
for i in search_second_level:
    start = time.time()
    looks = search(graph, i)
    end = time.time()
    elapsed = end-start
    print('Elapsed time is:', elapsed)
    level_list.append(2)
    looks_list.append(looks[0])
    names_list.append(i)
    times_list.append(elapsed)  
```

```python
third_level = list(graph.values())[1]
search_third_level = np.random.choice(third_level, 3, replace=False)
print("Searching for 3 names from same 2nd level person's 3rd level connections:")
print('Looking for', search_third_level)
for i in search_third_level:
    start = time.time()
    looks = search(graph, i)
    end = time.time()
    elapsed_3 = end-start
    print('Elapsed time is:', elapsed)
    level_list.append(3)
    looks_list.append(looks[0])
    names_list.append(i)
    times_list.append(elapsed)
```

```python
fourth_level = list(graph.values())[2]
search_fourth_level = np.random.choice(fourth_level, 4, replace=False)
print("Searching for 4 names from same 3rd level person's 4th level connections:")
print('Looking for', search_fourth_level)
for i in search_fourth_level:
    start = time.time()
    looks = search(graph, i)
    end = time.time()
    elapsed = end-start
    print('Elapsed time is:', elapsed)
    level_list.append(4)
    looks_list.append(looks[0])
    names_list.append(i)
    times_list.append(elapsed)
```

```python
fifth_level = list(graph.values())[3]
search_fifth_level = np.random.choice(fifth_level, 5, replace=False)
print("Searching for 5 names from same 4th level person's 5th level connections:")
print('Looking for', search_fifth_level)
for i in search_fifth_level:
    start = time.time()
    looks = search(graph, i)
    end = time.time()
    elapsed = end-start
    print('Elapsed time is:', elapsed)
    level_list.append(5)
    looks_list.append(looks[0])
    names_list.append(i)
    times_list.append(elapsed)
```

```python
data = {'Level': level_list,
       'Name': names_list,
       'Looks': looks_list,
       'Elapsed time': times_list}
df = pd.DataFrame(data)
df
```
